### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the data
df_market_data_scaled = scaler.fit_transform(df_market_data)

# Converting the scaled data back to a DataFrame
df_market_data_scaled = pd.DataFrame(df_market_data_scaled, columns=df_market_data.columns, index=df_market_data.index)

# Display the first few rows of the scaled data
df_market_data_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the data
df_market_data_scaled = scaler.fit_transform(df_market_data)

# Create a DataFrame with the scaled data
# Assuming 'coin_id' is the index of df_market_data, we'll retain it in the new DataFrame
df_scaled = pd.DataFrame(df_market_data_scaled, columns=df_market_data.columns, index=df_market_data.index)

# Display sample data from the scaled DataFrame
df_scaled.head()
```

### Find the Best Value for k Using the Original Data.

```python
# Create a list of k-values from 1 to 11
k_values = list(range(1, 12))
print(k_values)
```

```python
import warnings

# Suppress specific UserWarnings from sklearn.cluster
warnings.filterwarnings("ignore", category=UserWarning, module='sklearn.cluster')
```

```python
import os
from sklearn.cluster import KMeans

# Setting the environment variable to avoid the memory leak warning
os.environ['OMP_NUM_THREADS'] = '1' 

# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model using the current value of k
    # Explicitly setting n_init to avoid the future warning
    model = KMeans(n_clusters=k, n_init=10, random_state=0)

    # Fit the model to the scaled data
    model.fit(df_market_data_scaled)

    # Append the model's inertia to the inertia list
    inertia.append(model.inertia_)
```

```python
# Assuming 'k_values' is your list of k-values and 'inertia' is the list of inertia values

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k_values, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
import matplotlib.pyplot as plt

# Plotting the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(elbow_df['k'], elbow_df['inertia'], marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.

from sklearn.decomposition import PCA

# Create a PCA model instance with 3 components
pca = PCA(n_components=3)
```

```python
# Fit and transform the data using the PCA model
pca_data = pca.fit_transform(df_market_data_scaled)

# Create a DataFrame with the PCA data, using the index from the original DataFrame
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2', 'PC3'], index=df_market_data.index)

# View the first five rows of the PCA DataFrame
pca_df.head()

```

```python
# Retrieve the explained variance
explained_variance = pca.explained_variance_ratio_

# Display the explained variance
print("Explained Variance by Component: ", explained_variance)

# Display the cumulative explained variance
cumulative_explained_variance = explained_variance.cumsum()
print("Cumulative Explained Variance: ", cumulative_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.8950316570309841

```python
# Calculate the total explained variance of the first three principal components
total_explained_variance = sum(pca.explained_variance_ratio_[:3])

# Print the total explained variance
print("Total Explained Variance of the First Three Principal Components:", total_explained_variance)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
import holoviews as hv
from holoviews import opts

# Sample data for the Elbow curve (replace this with your actual data)
num_clusters = [2, 3, 4, 5, 6]
inertia_values_kmeans = [100, 80, 60, 40, 30]  # Inertia values for K-Means
inertia_values_optimized_kmeans = [90, 70, 50, 35, 25]  # Inertia values for Optimized K-Means

# Create the Elbow curve for K-Means
elbow_plot_kmeans = hv.Curve((num_clusters, inertia_values_kmeans), 'Number of Clusters', 'Inertia')

# Create the Elbow curve for Optimized K-Means
elbow_plot_optimized_kmeans = hv.Curve((num_clusters, inertia_values_optimized_kmeans), 'Number of Clusters', 'Inertia')

# Create a composite plot to contrast the Elbow curves
elbow_composite = (
    elbow_plot_kmeans
    * elbow_plot_optimized_kmeans
).opts(
    opts.Curve(width=400, height=300, title="Elbow Curve (K-Means)"),
    opts.Curve(width=400, height=300, title="Elbow Curve (Optimized K-Means)"),
    opts.Layout(title_format=""),
)

elbow_composite

```

```python
import holoviews as hv
from holoviews import opts

# Replace the following placeholders with your actual data and cluster labels
num_clusters = [2, 3, 4, 5, 6]
inertia_values_kmeans = [100, 80, 60, 40, 30]  # Inertia values for K-Means
inertia_values_optimized_kmeans = [90, 70, 50, 35, 25]  # Inertia values for Optimized K-Means
cluster_labels = [0, 1, 0, 1, 2]  # Cluster labels assigned to each data point

# Create a DataFrame with the data
import pandas as pd
data = pd.DataFrame({'Number of Clusters': num_clusters,
                     'Inertia (K-Means)': inertia_values_kmeans,
                     'Inertia (Optimized K-Means)': inertia_values_optimized_kmeans,
                     'Cluster': cluster_labels})

# Define a scatter plot for each cluster
scatter_plots = {cluster: hv.Scatter(data[data['Cluster'] == cluster], kdims=['Number of Clusters'], vdims=['Inertia (K-Means)', 'Inertia (Optimized K-Means)'])
                 for cluster in data['Cluster'].unique()}

# Create a composite plot to contrast the clusters
cluster_composite = hv.NdOverlay(scatter_plots).opts(
    opts.Scatter(width=400, height=300, marker='o', color=hv.Cycle('Category20'), title="Cluster Visualization"),
    opts.NdOverlay(legend_position='top_right')
)

cluster_composite
```
